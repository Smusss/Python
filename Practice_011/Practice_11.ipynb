```python
Дана функция f(x) = 5*x**2 + 10*x - 30

1. Определить корни+

2. Найти интервалы, на которых функция возрастает+

3. Найти интервалы, на которых функция убывает+

4. Построить график+

5. Вычислить вершину+

6. Определить промежутки, на котором f > 0+

7. Определить промежутки, на котором f < 0+
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b, c = 5, 10, -30
lim =  10
step = 0.01
```

```python
def func(x):
    return a*x**2 + b*x + c
```

```python
x = np.arange(-lim, lim, step)
```

```python
def take_roots(a, b, c):
    discr = (b**2 - 4*a*c)
    if discr > 0:
        x1 = (-b + discr**0.5)/(2*a)
        x2 = (-b - discr**0.5)/(2*a)
        return round(x1, 2), round (x2, 2)
    elif discr == 0:
        x = -b /(2*a)
        return (round(x, 2),)
    else:
        return (None,)
```

```python
roots = take_roots(a, b, c)
print (roots)
```

```python
min_y = (min(func(x)))
min_x = take_roots(a, b, c - min_y)[0]
print (min_x, min_y)
```

```python
x_down_plus = np.arange(-lim, min(roots) + step, step)
x_down_minus = np.arange(min(roots), min_x + step, step)
x_up_minus = np.arange(min_x, max(roots) + step, step)
x_up_plus = np.arange(max(roots), lim + step, step)
```

```python
plt.rcParams['lines.linestyle'] = '-'
plt.plot(x_down_plus, func(x_down_plus), 'b', label= 'Убывание больше 0')
plt.plot(x_up_plus, func(x_up_plus), 'r', label= 'Возрастание больше 0')
plt.rcParams['lines.linestyle'] = '-.'
plt.plot(x_down_minus, func(x_down_minus), 'b', label= 'Убывание меньше 0')
plt.plot(x_up_minus, func(x_up_minus), 'r', label= 'Возрастание меньшне 0')

if len (roots) == 2:
    plt.plot(roots[0], 0, 'gx', label= f'Корни ({roots[0]}, {roots [1]})')
    plt.plot(roots[1], 0, 'gx')
else:
    if roots[0] is not None:
        plt.plot(roots[0], 0, 'gx', label= f'Корни ({roots[0]}, {roots [1]})')
plt.plot(min_x, min_y, 'yo', label=  f'Экстремум функции ({min_x}, {min_y})')
plt.legend()
plt.grid()
```
