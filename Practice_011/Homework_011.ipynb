```python
import numpy as np
import matplotlib.pyplot as plt
#from sympy import symbols, solve
```

```python
a, b, c, d, e = -12, -18, 5, 10, -30
koef = [a, b, c, d, e]
lim = 11
step = 0.01
```

```python
x_arr = np.arange(-lim, lim + step, step)
print(x_arr[5])
```

```python
def func(x):
    return a*x**4*np.sin(np.cos(x)) + b*x**3 + c*x**2 + d*x + e
```

```python
def take_roots(k_array:list):
    all_roots = np.roots(k_array)
    #print (all_roots)
    roots = []
    for r in all_roots: roots.append(round(r.real,2))
    return roots 
```

```python
roots = take_roots(koef)
print(roots)
```

```python
min_y = round(min(func(x_arr)),2)
print(min_y)
k_array_min_x = koef
k_array_min_x[-1] = round(koef[-1] - min_y,2)
min_x = take_roots(k_array_min_x)[3]
print(min_x)
```

```python
dict = {}
print(x_arr[0:10])
for i in range(len(x_arr)):
    if i == 0:
        if func(x_arr[i]) >= 0:
            dict[round(x_arr[i],2)] = 'pu'
        else:
            dict[round(x_arr[i],2)] = 'mu'
            minus_up.append(round(x_arr[i],2))
    else:
        if func(x_arr[i]) >= func(x_arr[i-1]):
            if func(x_arr[i]) >= 0:
                dict[round(x_arr[i],2)] = 'pu'
            else:
                dict[round(x_arr[i],2)] = 'mu'
        else:
            if func(x_arr[i]) >= 0:
                dict[round(x_arr[i],2)] = 'pd'
            else:
                dict[round(x_arr[i],2)] = 'md'
```

```python
plt.title('The part of function')
plt.rcParams['lines.linestyle'] = '-'

plt.plot(roots[0], 0, 'gx', label= f'Корни ({roots[0]}, {roots[2]})')
plt.plot(roots[2], 0, 'gx')

plt.plot(min_x, min_y,'ro', label= f'Экстремум ({min_x, min_y})')
plt.plot(x_arr, func(x_arr))
plt.legend()
plt.grid()
```
